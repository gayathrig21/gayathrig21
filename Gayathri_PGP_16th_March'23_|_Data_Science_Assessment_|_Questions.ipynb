- Given The **Dataset - 1**

- Calculate the **Percentage Month-over-Month** change in **Revenue.**

  - **Percentage change**, rounded to the **2nd decimal point**

## **Question 2:**

**Background:** 
- Some forecasting methods are extremely simple and surprisingly effective.  

- Nave forecast is one of them, we simply set all forecasts to be the value of the last observation.  

- Our goal is to develop a nave forecast for a new metric called "distance per dollar" defined as:  

    
    (distance_to_travel/monetary_cost)

**Problem:**

- To develop this forecast,  sum **"distance to travel"  and "monetary cost"** values at a **monthly level** before calculating **"distance per dollar".** 

- This value becomes your **actual value for the current month.** 

- The next step is to populate the **forecasted value** for each month. This can be achieved simply by getting the previous month's value in a separate column. 

- Now, we have **actual and forecasted values.** This is your **nave forecast.** 

- Lets evaluate our model by calculating an error matrix called **Root Mean Squared Error (RMSE).** 

**Note:**

RMSE is defined as 

    sqrt(mean(square(actual - forecast))

Report out the RMSE rounded to the 2nd decimal spot.

```python
df1_uber['Request_Date']=pd.to_datetime(df1_uber['Request_Date']).dt.strftime('%Y-%m')
```

```python
df1_uber.set_index('Request_Date', inplace=True)
```

```python
df1_uber=df1_uber.groupby('Request_Date').sum()
```

```python
df1_uber['Actual_Distance_per_Dollar']=df1_uber['Distance_To_Travel']/df1_uber['Value']
```

```python
df1_uber=df1_uber.drop(['Distance_To_Travel',	'Value'], axis=1)
```

```python
df1_uber
```

```python
df2_uber=df1_uber.shift(periods=1) 
```

```python
df2_uber
```

```python
df3_uber=pd.merge(df1_uber, df2_uber, on='Request_Date')
```

```python
df3_uber.rename(columns = {'Actual_Distance_per_Dollar_x':'Actual_Distance_per_Dollar','Actual_Distance_per_Dollar_y':'Forecasted_Distance_per_Dollar'}, inplace = True)
```

```python
df3_uber
```

```python
rmse= np.sqrt(np.mean(np.square(df3_uber['Actual_Distance_per_Dollar'] - df3_uber['Forecasted_Distance_per_Dollar'])))
```

```python
rmse
```
